По поводу выбросов и пропущенных значений:
- "явные" (грубые) выбросы можно отсеивать, применяя различные критерии, но, во-первых, почти всегда можно найти "правильный" критерий, который покажет, что это не выброс, а, во-вторых, при такой оценке нужно использовать знания из предметной области, а не только из статистики и теории вероятостей. К тому же большинство критериев ориентируется на нормальные или близкие к нему распределения, что в рассмотренных случаях не очень соблюдалось. 
- замена пропущенных значений чем-либо - очень тонкая вещь, которая без предметных знаний просто опасна. Можно оценить значимость влияния параметров на что-то изучаемое, и, если у каких-то мало значащих параметров много пропусков, то их можно исключить из рассмотрения. Любая замена приводит к искажению результатов.

```python
# смотрим, как влияет количество не пустых значений на количество остающихся строк на отобранных параметрах
df2 = df[['surgery?', 'Age', 'Hospital Number', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities','capillary refill time']]
#df2
df2.info()
for i in range(1, len(df2.axes[1])+1):
    print('если в строке по меньшей мере', i, 'не пустое значение (', len(df2.axes[1]) - i,'пропусков), то остается записей', len(df2.dropna(thresh = i).axes[0]))
```

На моем наборе данных полностью заполнены 164 строки (чуть больше половины). Из параметров хуже всего с "respiratory rate" (частотой дыхания) и "temperature of extremities" (температурой конечностей) - заполнено порядка 80% строк.

```python
# Заменим в ранговых параметрах ('surgery?', 'temperature of extremities', 'capillary refill time') отсутствующие значения
# на константу 0 - отсутствие данных (там нет такого ранга)
# а 'rectal temperature', 'pulse' и 'respiratory rate' на медиану по группам ранговых переменных
df2['surgery?'].fillna(value = 0.0, inplace = True)
df2['temperature of extremities'].fillna(value = 0.0, inplace = True)
df2['capillary refill time'].fillna(value = 0.0, inplace = True)

#print(df2.groupby(['surgery?', 'temperature of extremities', 'capillary refill time'])['rectal temperature'].median())

df2['rectal temperature'].fillna(value = df2.groupby(['surgery?', 'temperature of extremities', 'capillary refill time'])['rectal temperature'].transform('median').round(1), inplace=True)
df2['pulse'].fillna(value = df2.groupby(['surgery?', 'temperature of extremities', 'capillary refill time'])['pulse'].transform('median').round(1), inplace=True)
df2['respiratory rate'].fillna(value = df2.groupby(['surgery?', 'temperature of extremities', 'capillary refill time'])['respiratory rate'].transform('median').round(1), inplace=True)

# а незамененных - на их моды
df2['rectal temperature'].fillna(value = df2['rectal temperature'].mode()[0], inplace=True)
df2['pulse'].fillna(value = df2['pulse'].mode()[0], inplace=True)
df2['respiratory rate'].fillna(value = df2['respiratory rate'].mode()[0], inplace=True)

df2.info()
df2.describe()
```
